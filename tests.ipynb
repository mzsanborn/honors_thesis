Figure 8

```python
dim = 1000
simulations = np.empty((dim), dtype=object)
vectors = [] 
char_lengths = []
new_vector = np.empty((dim), dtype=object)
new_idxs = [] 
mu = 10 ** (-7)
r= 0.05
for m in [0.025, 0.05, 0.1, 0.2]: 
    for N in [200, 600, 1000]:
        for s in [0.08, 0.16, 0.24, 0.32, 0.4]:
            variance = mu * (1-mu)
            std = math.sqrt(m * (1-m))
            char_length = ((variance * std) / (N * mu * math.sqrt(2*s) * 2)) ** (0.5)
            for i in range(dim):
                new_vector[i] = {0: N}
            new_idx =1
            vector, new_idx = linear_simulation(new_vector, new_idx, 10000, dim, N, s, r, m, mu)
            new_idxs.append(copy.deepcopy(new_idx))
            vectors.append(copy.deepcopy(vector))
            char_lengths.append(copy.deepcopy(char_length))
```

```python
plot_graph8(vectors, new_idxs, char_lengths)
```
